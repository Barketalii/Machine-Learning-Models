**LinearRegression**
It‚Äôs a machine learning model from Scikit-learn (sklearn).
Linear regression is used for predicting a continuous target variable based on input features.
Example: predicting house prices based on size, location, number of rooms.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([[1],[2],[3],[4],[5]])
y=np.array([[1],[4],[6],[8],[10]])
```

```python
import seaborn as sns
sns.scatterplot(X)
```

When we write LinearRegression(), we are creating an instance (object) of that class.

model is the variable that holds this object.

```python
model=LinearRegression()
```

**.fit**
Training phase: the model learns patterns from X and y.

Internally, it calculates weights, coefficients, or structures (depending on the algorithm).

After .fit(), the model is ready to make predictions.

```python
model.fit(X,y)
```

```python
new_input=np.array([[6]])
```

```python
pred_output=model.predict(new_input)
```
